The table below is my inital data set and it is separated into title and another list title "String List Data" with more data I will access further down

```python
likesTable = pd.DataFrame(data["likes_media_likes"])
```

```python
likesTable.head()
```

Below is a list comprehension used to help me access the timestamp data from within the dictionary titled "string_list_data". I took the data from this and placed it into a seperate column in the table.

```python
timeStampData = [x["string_list_data"][0]["timestamp"] for x in data["likes_media_likes"]]
```

```python
timeStampTab = pd.DataFrame(timeStampData)
```

```python
likesTable["Timestamps"] = timeStampTab
```

```python
likesTable
```

Below I am grouping the information from my table above by account and sorting the values to see which accounts have the most instances or which accounts I have liked the most posts from.

```python
groupedLikes = likesTable.groupby("title").count().sort_values("string_list_data", ascending = False)
```

```python
groupedLikes.head(50)
```
